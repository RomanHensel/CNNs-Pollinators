```python
import shutil, random, os

random.seed(42)

classes = ["background", "bee", "bumblebee", "hoverfly", "moth"]

# Create destination folders
for split in ["train", "validation", "test"]:
    for cls in classes:
        os.makedirs(f"/content/data/{split}/{cls}", exist_ok=True)

# Split and copy
for cls in classes:
    src   = os.path.join(dataset_path, cls)
    files = [f for f in os.listdir(src) if not f.startswith(".")]  # skip .DS_Store
    random.shuffle(files)
    n = len(files)

    n_train = int(n * 0.70)
    n_val   = int(n * 0.15)

    train_files = files[:n_train]
    val_files   = files[n_train:n_train+n_val]
    test_files  = files[n_train+n_val:]

    for f in train_files:
        shutil.copy(os.path.join(src, f), f"/content/data/train/{cls}/{f}")
    for f in val_files:
        shutil.copy(os.path.join(src, f), f"/content/data/validation/{cls}/{f}")
    for f in test_files:
        shutil.copy(os.path.join(src, f), f"/content/data/test/{cls}/{f}")

    print(f"{cls} → {n_train} train | {n_val} val | {len(test_files)} test")
```

```python
img_size   = (256, 256)
batch_size = 32

train_ds = keras.utils.image_dataset_from_directory(
    "/content/data/train",
    image_size = img_size,
    batch_size = batch_size,
    label_mode = "categorical",
    shuffle    = True
)

val_ds = keras.utils.image_dataset_from_directory(
    "/content/data/validation",
    image_size = img_size,
    batch_size = batch_size,
    label_mode = "categorical",
    shuffle    = False
)

print("Classes:", train_ds.class_names)
```

```python
class_counts = {
    "background": 889,
    "bee":        8332,
    "bumblebee":  299,
    "hoverfly":   1386,
    "moth":       207
}

total     = sum(class_counts.values())
n_classes = len(class_counts)

# Alphabetical order: background=0, bee=1, bumblebee=2, hoverfly=3, moth=4
class_weights = {
    i: total / (n_classes * count)
    for i, (cls, count) in enumerate(sorted(class_counts.items()))
}

print(class_weights)
```

```python
base_model = MobileNetV2(
    weights     = "imagenet",
    include_top = False,
    input_shape = (256, 256, 3)
)
base_model.trainable = False

model_256 = keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(5, activation="softmax")
])

model_256.compile(
    optimizer = keras.optimizers.Adam(learning_rate=0.001),
    loss      = "categorical_crossentropy",
    metrics   = ["accuracy"]
)

model_256.summary()
```

```python
callbacks = [
    keras.callbacks.ModelCheckpoint(
        "/content/drive/MyDrive/best_model_256.keras",
        save_best_only = True,
        monitor        = "val_accuracy"
    ),
    keras.callbacks.EarlyStopping(
        patience             = 5,
        monitor              = "val_accuracy",
        restore_best_weights = True
    )
]

history_256 = model_256.fit(
    train_ds,
    epochs          = 20,
    validation_data = val_ds,
    class_weight    = class_weights,
    callbacks       = callbacks
)
```

```python
# Unfreeze the last 30 layers of the base model
base_model.trainable = True
for layer in base_model.layers[:-30]:
    layer.trainable = False  # keep early layers frozen

# Recompile with a much lower learning rate
model_256.compile(
    optimizer = keras.optimizers.Adam(learning_rate=0.0001),  # 10x lower!
    loss      = "categorical_crossentropy",
    metrics   = ["accuracy"]
)

# Continue training
history_finetune = model_256.fit(
    train_ds,
    epochs          = 20,
    validation_data = val_ds,
    class_weight    = class_weights,
    callbacks = [
        keras.callbacks.ModelCheckpoint(
            "/content/drive/MyDrive/best_model_256.keras",
            save_best_only = True,
            monitor        = "val_accuracy"
        ),
        keras.callbacks.EarlyStopping(
            patience             = 7,
            monitor              = "val_accuracy",
            restore_best_weights = True
        )
    ]
)
```

```python
# Step 1: Add augmentation to training data
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])

train_ds_aug = train_ds.map(
    lambda x, y: (data_augmentation(x, training=True), y)
).prefetch(tf.data.AUTOTUNE)

val_ds_pre = val_ds.prefetch(tf.data.AUTOTUNE)

# Step 2: Normalize pixel values
normalization = layers.Rescaling(1./255)

# Step 3: Build fresh model with normalization + augmentation built in
base_model2 = MobileNetV2(
    weights     = "imagenet",
    include_top = False,
    input_shape = (256, 256, 3)
)
base_model2.trainable = True
for layer in base_model2.layers[:-30]:
    layer.trainable = False

inputs = keras.Input(shape=(256, 256, 3))
x = normalization(inputs)
x = base_model2(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(0.4)(x)          # slightly higher dropout
outputs = layers.Dense(5, activation="softmax")(x)

model_256_v2 = keras.Model(inputs, outputs)

model_256_v2.compile(
    optimizer = keras.optimizers.Adam(learning_rate=0.0001),
    loss      = "categorical_crossentropy",
    metrics   = ["accuracy"]
)

# Step 4: Train
history_256_v2 = model_256_v2.fit(
    train_ds_aug,
    epochs          = 30,
    validation_data = val_ds_pre,
    class_weight    = class_weights,
    callbacks = [
        keras.callbacks.ModelCheckpoint(
            "/content/drive/MyDrive/best_model_256_v2.keras",
            save_best_only = True,
            monitor        = "val_accuracy"
        ),
        keras.callbacks.EarlyStopping(
            patience             = 7,
            monitor              = "val_accuracy",
            restore_best_weights = True
        ),
        keras.callbacks.ReduceLROnPlateau(   # automatically reduces learning rate
            monitor  = "val_loss",           # when val_loss stops improving
            factor   = 0.5,
            patience = 3,
            verbose  = 1
        )
    ]
)
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns

# ── 1. Settings ───────────────────────────────────────────────────────────────
MODEL_PATH  = "/content/drive/MyDrive/best_model_256_v2.keras"
DATA_ROOT   = "/content/drive/MyDrive/BuzzSet_Edited"
IMG_SIZE    = 256
BATCH_SIZE  = 32

# Moth excluded — only 3 classes now
CLASS_NAMES = ["bee", "bumblebee", "hoverfly"]   # ← adjust order to match training!

# ── 2. Helper: load images from one environment ───────────────────────────────
def load_environment(env_name):
    paths, labels = [], []
    for class_idx, cls in enumerate(CLASS_NAMES):
        folder = os.path.join(DATA_ROOT, cls, env_name)
        if not os.path.isdir(folder):
            print(f"  Warning: {folder} not found, skipping")
            continue
        for fname in os.listdir(folder):
            if fname.lower().endswith((".jpg", ".jpeg", ".png")):
                paths.append(os.path.join(folder, fname))
                labels.append(class_idx)
    print(f"  [{env_name}] {len(paths)} images across {len(set(labels))} classes")
    return paths, labels

def make_dataset(paths, labels):
    def read_image(path, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_image(img, channels=3, expand_animations=False)
        img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
        return img, label
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(read_image, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    return ds

# ── 3. Load both environments ─────────────────────────────────────────────────
print("Loading Phacelia...")
phacelia_paths, phacelia_labels = load_environment("Phacelia")

print("Loading Others...")
others_paths, others_labels = load_environment("Others")

# ── 4. Compute sample weights ─────────────────────────────────────────────────
# This makes each class count equally regardless of how many images it has.
# We compute weights separately per environment so each is fairly evaluated
# on its own internal balance.

def get_sample_weights(labels):
    labels = np.array(labels)
    classes = np.unique(labels)
    cw = compute_class_weight("balanced", classes=classes, y=labels)
    weight_map = dict(zip(classes, cw))
    return np.array([weight_map[l] for l in labels])

phacelia_weights = get_sample_weights(phacelia_labels)
others_weights   = get_sample_weights(others_labels)

print("\nPhacelia class weights:")
for i, name in enumerate(CLASS_NAMES):
    w = dict(zip(*np.unique(phacelia_labels, return_counts=True)))
    print(f"  {name}: {phacelia_weights[phacelia_labels.index(i) if i in phacelia_labels else 0]:.2f}")

# ── 5. Load model & predict ───────────────────────────────────────────────────
print("\nLoading model...")
model = keras.models.load_model(MODEL_PATH)
print("Model loaded ✓")

print("\nPredicting on Phacelia...")
phacelia_ds   = make_dataset(phacelia_paths, phacelia_labels)
phacelia_probs = model.predict(phacelia_ds, verbose=1)
phacelia_preds = np.argmax(phacelia_probs, axis=1)

print("\nPredicting on Others...")
others_ds   = make_dataset(others_paths, others_labels)
others_probs = model.predict(others_ds, verbose=1)
others_preds = np.argmax(others_probs, axis=1)

# ── 6. Weighted accuracy + classification report ──────────────────────────────
def weighted_accuracy(true, preds, weights):
    true = np.array(true)
    correct = (true == preds).astype(float)
    return np.sum(correct * weights) / np.sum(weights)

phacelia_true = np.array(phacelia_labels)
others_true   = np.array(others_labels)

acc_p = weighted_accuracy(phacelia_true, phacelia_preds, phacelia_weights)
acc_o = weighted_accuracy(others_true,   others_preds,   others_weights)

for name, true, preds, weights in [
    ("Phacelia", phacelia_true, phacelia_preds, phacelia_weights),
    ("Others",   others_true,   others_preds,   others_weights),
]:
    print(f"\n{'='*52}")
    print(f"  {name}  —  weighted accuracy: {weighted_accuracy(true, preds, weights)*100:.1f}%")
    print(f"{'='*52}")
    print(classification_report(
        true, preds,
        target_names=CLASS_NAMES,
        sample_weight=weights,   # ← weights applied here too
        digits=3
    ))

print(f"\nWeighted accuracy difference (Phacelia − Others): {(acc_p - acc_o)*100:+.1f} pp")

# ── 7. Confusion matrices (normalised, weighted) ───────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for ax, true, preds, weights, name in zip(
    axes,
    [phacelia_true, others_true],
    [phacelia_preds, others_preds],
    [phacelia_weights, others_weights],
    ["Phacelia", "Others"]
):
    cm = confusion_matrix(true, preds, sample_weight=weights)
    cm_pct = cm / cm.sum(axis=1, keepdims=True)
    sns.heatmap(
        cm_pct, annot=True, fmt=".0%", cmap="Blues",
        xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES,
        ax=ax, vmin=0, vmax=1
    )
    acc = weighted_accuracy(true, preds, weights)
    ax.set_title(f"{name}  (weighted acc = {acc*100:.1f}%)", fontsize=12)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")

plt.suptitle("Pollinator ID by environment (class-balanced evaluation)", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("/content/drive/MyDrive/environment_comparison_weighted.png", dpi=150)
plt.show()
print("Saved ✓")
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns

# ── 1. Settings ───────────────────────────────────────────────────────────────
MODEL_PATH    = "/content/drive/MyDrive/best_model_256_v2.keras"
DATA_ROOT     = "/content/drive/MyDrive/BuzzSet_Edited"
IMG_SIZE      = 256
BATCH_SIZE    = 32

ALL_CLASSES   = ["background", "bee", "bumblebee", "hoverfly", "moth"]
CLASS_NAMES   = ["bee", "bumblebee", "hoverfly"]
VALID_INDICES = [1, 2, 3]  # bee=1, bumblebee=2, hoverfly=3

# ── 2. Helper: load images from one environment ───────────────────────────────
def load_environment(env_name):
    paths, labels = [], []
    for class_idx, cls in enumerate(CLASS_NAMES):
        real_idx = VALID_INDICES[class_idx]
        folder = os.path.join(DATA_ROOT, cls, env_name)
        if not os.path.isdir(folder):
            print(f"  Warning: {folder} not found, skipping")
            continue
        for fname in os.listdir(folder):
            if fname.lower().endswith((".jpg", ".jpeg", ".png")):
                paths.append(os.path.join(folder, fname))
                labels.append(real_idx)
    print(f"  [{env_name}] {len(paths)} images across {len(set(labels))} classes")
    return paths, labels

def make_dataset(paths, labels):
    def read_image(path, label):
        img = tf.io.read_file(path)
        img = tf.image.decode_image(img, channels=3, expand_animations=False)
        img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
        return img, label
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(read_image, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    return ds

# ── 3. Load both environments ─────────────────────────────────────────────────
print("Loading Phacelia...")
phacelia_paths, phacelia_labels = load_environment("Phacelia")

print("Loading Others...")
others_paths, others_labels = load_environment("Others")

# ── 4. Compute sample weights ─────────────────────────────────────────────────
def get_sample_weights(labels):
    labels = np.array(labels)
    classes = np.unique(labels)
    cw = compute_class_weight("balanced", classes=classes, y=labels)
    weight_map = dict(zip(classes, cw))
    return np.array([weight_map[l] for l in labels])

phacelia_weights = get_sample_weights(phacelia_labels)
others_weights   = get_sample_weights(others_labels)

print("\nPhacelia class weights:")
for name, idx in zip(CLASS_NAMES, VALID_INDICES):
    mask = np.array(phacelia_labels) == idx
    if mask.any():
        print(f"  {name}: {phacelia_weights[mask][0]:.2f}")

print("\nOthers class weights:")
for name, idx in zip(CLASS_NAMES, VALID_INDICES):
    mask = np.array(others_labels) == idx
    if mask.any():
        print(f"  {name}: {others_weights[mask][0]:.2f}")

# ── 5. Load model & predict ───────────────────────────────────────────────────
print("\nLoading model...")
model = keras.models.load_model(MODEL_PATH)
print("Model loaded ✓")

print("\nPredicting on Phacelia...")
phacelia_ds    = make_dataset(phacelia_paths, phacelia_labels)
phacelia_probs = model.predict(phacelia_ds, verbose=1)

print("\nPredicting on Others...")
others_ds    = make_dataset(others_paths, others_labels)
others_probs = model.predict(others_ds, verbose=1)

# ── 6. Filter predictions to valid classes only ───────────────────────────────
def filter_to_valid(labels, probs, valid_indices):
    valid_probs = probs[:, valid_indices]
    new_preds   = np.array(valid_indices)[np.argmax(valid_probs, axis=1)]
    return np.array(labels), new_preds

phacelia_true, phacelia_preds = filter_to_valid(
    phacelia_labels, phacelia_probs, VALID_INDICES)

others_true, others_preds = filter_to_valid(
    others_labels, others_probs, VALID_INDICES)

# ── 7. Weighted accuracy ──────────────────────────────────────────────────────
def weighted_accuracy(true, preds, weights):
    correct = (true == preds).astype(float)
    return np.sum(correct * weights) / np.sum(weights)

acc_p = weighted_accuracy(phacelia_true, phacelia_preds, phacelia_weights)
acc_o = weighted_accuracy(others_true,   others_preds,   others_weights)

# ── 8. Print reports ──────────────────────────────────────────────────────────
for name, true, preds, weights in [
    ("Phacelia", phacelia_true, phacelia_preds, phacelia_weights),
    ("Others",   others_true,   others_preds,   others_weights),
]:
    print(f"\n{'='*52}")
    print(f"  {name}  —  weighted accuracy: {weighted_accuracy(true, preds, weights)*100:.1f}%")
    print(f"{'='*52}")
    print(classification_report(
        true, preds,
        labels=VALID_INDICES,
        target_names=CLASS_NAMES,
        sample_weight=weights,
        digits=3
    ))

print(f"\nWeighted accuracy difference (Phacelia − Others): {(acc_p - acc_o)*100:+.1f} percentage points")

# ── 9. Confusion matrices ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for ax, true, preds, weights, name in zip(
    axes,
    [phacelia_true, others_true],
    [phacelia_preds, others_preds],
    [phacelia_weights, others_weights],
    ["Phacelia", "Others"]
):
    cm = confusion_matrix(true, preds, labels=VALID_INDICES, sample_weight=weights)
    cm_pct = cm / cm.sum(axis=1, keepdims=True)
    sns.heatmap(
        cm_pct, annot=True, fmt=".0%", cmap="Blues",
        xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES,
        ax=ax, vmin=0, vmax=1
    )
    acc = weighted_accuracy(true, preds, weights)
    ax.set_title(f"{name}  (weighted acc = {acc*100:.1f}%)", fontsize=12)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")

plt.suptitle("Pollinator identification by environment (class-balanced)", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("/content/drive/MyDrive/environment_comparison_weighted.png", dpi=150)
plt.show()
print("Saved to Drive ✓")
```
