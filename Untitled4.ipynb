```python
import shutil, random, os
from pathlib import Path

random.seed(42)

pollinators = ["bee", "bumblebee", "hoverfly"]
backgrounds = ["Phacelia", "Others"]

base_out = "/content/data_background"

# Vorherige Daten löschen, falls vorhanden (frischer Start)
if os.path.exists(base_out):
    shutil.rmtree(base_out)

# Zielordner anlegen
for split in ["train", "val"]:
    for p in pollinators:
        os.makedirs(f"{base_out}/{split}/{p}", exist_ok=True)

for bg in backgrounds:
    for p in pollinators:
        os.makedirs(f"{base_out}/test_{bg.lower()}/{p}", exist_ok=True)

# Split-Funktion mit anpassbaren Anteilen
def split_and_copy(pollinator, background, train=0.70, val=0.15, test=0.15):
    src = os.path.join(bg_dataset_path, pollinator, background)
    files = [f for f in os.listdir(src) if not f.startswith(".")]
    random.shuffle(files)
    n = len(files)

    n_train = int(n * train)
    n_val   = int(n * val)

    train_files = files[:n_train]
    val_files   = files[n_train:n_train+n_val]
    test_files  = files[n_train+n_val:]

    for f in train_files:
        shutil.copy(os.path.join(src, f), f"{base_out}/train/{pollinator}/{f}")
    for f in val_files:
        shutil.copy(os.path.join(src, f), f"{base_out}/val/{pollinator}/{f}")

    test_folder = f"{base_out}/test_{background.lower()}/{pollinator}"
    for f in test_files:
        shutil.copy(os.path.join(src, f), f"{test_folder}/{f}")

    print(f"{pollinator}/{background} → {n_train} train | {n_val} val | {len(test_files)} test")

# Standard-Split für bee und hoverfly
for p in ["bee", "hoverfly"]:
    for bg in backgrounds:
        split_and_copy(p, bg)

# Angepasster Split NUR für bumblebee (mehr Testdaten)
for bg in backgrounds:
    split_and_copy("bumblebee", bg, train=0.60, val=0.10, test=0.30)
```

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torchvision.models import resnet18, ResNet18_Weights
from torch.utils.data import DataLoader
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.metrics import f1_score, confusion_matrix, classification_report
import os
from time import perf_counter

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

train_dir = f"{base_out}/train"
val_dir   = f"{base_out}/val"
test_phacelia_dir = f"{base_out}/test_phacelia"
test_others_dir   = f"{base_out}/test_others"

output_dir = Path("/content/drive/MyDrive/Pollinators_Background")
output_dir.mkdir(parents=True, exist_ok=True)

# Hyperparameter
img_size = 256
batch_size = 32
num_epochs = 10
patience = 3
min_delta = 0.001
lr = 1e-4

transform = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                          [0.229, 0.224, 0.225])
])

train_dataset = datasets.ImageFolder(train_dir, transform=transform)
val_dataset   = datasets.ImageFolder(val_dir, transform=transform)
test_phacelia_dataset = datasets.ImageFolder(test_phacelia_dir, transform=transform)
test_others_dataset   = datasets.ImageFolder(test_others_dir, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_phacelia_loader = DataLoader(test_phacelia_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_others_loader   = DataLoader(test_others_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

print("Klassen:", train_dataset.classes)
print("Train:", len(train_dataset), "| Val:", len(val_dataset))
print("Test Phacelia:", len(test_phacelia_dataset), "| Test Others:", len(test_others_dataset))
```

```python
# Class Weights berechnen (automatisch aus den Trainingsdaten)
class_counts = torch.bincount(torch.tensor(train_dataset.targets))
class_weights = 1.0 / class_counts.float()
class_weights = class_weights / class_weights.sum() * len(class_counts)
class_weights = class_weights.to(device)

print("Klassen-Reihenfolge:", train_dataset.classes)
print("Klassen-Counts:", class_counts)
print("Klassen-Weights:", class_weights)

# ResNet18 mit vortrainierten Gewichten laden
model = resnet18(weights=ResNet18_Weights.DEFAULT)
model.fc = nn.Linear(model.fc.in_features, len(train_dataset.classes))
model = model.to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

print("\nModell bereit.")
```

```python
def run_epoch(loader, model, criterion, optimizer=None, train=True):
    if train:
        model.train()
    else:
        model.eval()

    running_loss = 0.0
    running_correct = 0
    running_total = 0

    for imgs, labels in loader:
        imgs = imgs.to(device)
        labels = labels.to(device)

        if train:
            optimizer.zero_grad()

        with torch.set_grad_enabled(train):
            outputs = model(imgs)
            loss = criterion(outputs, labels)

            if train:
                loss.backward()
                optimizer.step()

        running_loss += loss.item()
        preds = outputs.argmax(dim=1)
        running_correct += (preds == labels).sum().item()
        running_total += labels.size(0)

    avg_loss = running_loss / len(loader)
    avg_acc = running_correct / running_total * 100
    return avg_loss, avg_acc
```

```python
best_val_loss = float("inf")
best_epoch = 0
counter = 0
best_path = output_dir / "best_model_resnet18_background.pth"

start_time = perf_counter()

for epoch in range(num_epochs):
    train_loss, train_acc = run_epoch(train_loader, model, criterion, optimizer, train=True)
    val_loss, val_acc = run_epoch(val_loader, model, criterion, train=False)

    if val_loss < best_val_loss - min_delta:
        best_val_loss = val_loss
        best_epoch = epoch + 1
        counter = 0
        torch.save(model.state_dict(), best_path)
    else:
        counter += 1

    print(
        f"Epoch {epoch+1}/{num_epochs} | "
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )

    if counter >= patience:
        print(f"Early stopping at epoch {epoch+1}")
        break

elapsed_time = perf_counter() - start_time
print(f"\nTrainingszeit: {elapsed_time:.1f}s")

# Bestes Modell laden
model.load_state_dict(torch.load(best_path, map_location=device))
model = model.to(device)
```

```python
def evaluate_test(loader, dataset, model, criterion, class_names, background_name):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device)
            labels = labels.to(device)

            outputs = model(imgs)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            preds = outputs.argmax(dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    test_loss = running_loss / len(loader)
    test_acc = (np.array(all_preds) == np.array(all_labels)).mean() * 100
    macro_f1 = f1_score(all_labels, all_preds, average="macro")
    cm = confusion_matrix(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0)

    print(f"\n=== Test-Ergebnis: {background_name} ===")
    print(f"Test Accuracy: {test_acc:.2f}%")
    print(f"Macro F1-Score: {macro_f1:.4f}")
    print(f"\nConfusion Matrix:\n{pd.DataFrame(cm, index=class_names, columns=class_names)}")
    print(f"\nClassification Report:\n{report}")

    return test_loss, test_acc, macro_f1, cm, report
```

```python
class_names = train_dataset.classes

# Test auf Phacelia
loss_p, acc_p, f1_p, cm_p, report_p = evaluate_test(
    test_phacelia_loader, test_phacelia_dataset, model, criterion, class_names, "Phacelia"
)

# Test auf Others
loss_o, acc_o, f1_o, cm_o, report_o = evaluate_test(
    test_others_loader, test_others_dataset, model, criterion, class_names, "Others"
)

# Direkter Vergleich
print("\n" + "="*50)
print("VERGLEICH")
print("="*50)
print(f"Phacelia → Accuracy: {acc_p:.2f}% | Macro F1: {f1_p:.4f}")
print(f"Others   → Accuracy: {acc_o:.2f}% | Macro F1: {f1_o:.4f}")
print(f"Differenz: {acc_p - acc_o:.2f} Prozentpunkte")

# Ergebnisse speichern
results_summary = pd.DataFrame([
    {"background": "phacelia", "test_acc": acc_p, "macro_f1": f1_p, "n_test": len(test_phacelia_dataset)},
    {"background": "others",   "test_acc": acc_o, "macro_f1": f1_o, "n_test": len(test_others_dataset)}
])
results_summary.to_csv(output_dir / "background_comparison_results.csv", index=False)

pd.DataFrame(cm_p, index=class_names, columns=class_names).to_csv(output_dir / "confusion_matrix_phacelia.csv")
pd.DataFrame(cm_o, index=class_names, columns=class_names).to_csv(output_dir / "confusion_matrix_others.csv")

print("\nErgebnisse gespeichert in:", output_dir)
```

```python
# Prüfen ob irgendein Dateiname in mehreren Splits vorkommt
import os

def get_filenames(folder):
    names = set()
    for root, dirs, files in os.walk(folder):
        for f in files:
            if not f.startswith("."):
                names.add(f)
    return names

train_files = get_filenames(f"{base_out}/train")
val_files = get_filenames(f"{base_out}/val")
test_p_files = get_filenames(f"{base_out}/test_phacelia")
test_o_files = get_filenames(f"{base_out}/test_others")

overlap_train_testp = train_files & test_p_files
overlap_train_testo = train_files & test_o_files
overlap_val_test = val_files & (test_p_files | test_o_files)

print("Überlappung train ↔ test_phacelia:", len(overlap_train_testp))
print("Überlappung train ↔ test_others:", len(overlap_train_testo))
print("Überlappung val ↔ test:", len(overlap_val_test))
```

```python
def evaluate_test(loader, dataset, model, criterion, class_names, background_name):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device)
            labels = labels.to(device)
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            running_loss += loss.item()
            preds = outputs.argmax(dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    test_loss = running_loss / len(loader)
    test_acc = (np.array(all_preds) == np.array(all_labels)).mean() * 100
    macro_f1 = f1_score(all_labels, all_preds, average="macro")
    f1_per_class = f1_score(all_labels, all_preds, average=None)
    cm = confusion_matrix(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0)

    return test_loss, test_acc, macro_f1, f1_per_class, cm, report

loss_p, acc_p, f1_p, f1_per_class_p, cm_p, report_p = evaluate_test(
    test_phacelia_loader, test_phacelia_dataset, model, criterion, class_names, "Phacelia")
loss_o, acc_o, f1_o, f1_per_class_o, cm_o, report_o = evaluate_test(
    test_others_loader, test_others_dataset, model, criterion, class_names, "Others")
```

```python
from torchvision.models import mobilenet_v2, MobileNet_V2_Weights

# Class Weights sind schon aus Zelle 5 vorhanden (class_weights, gleicher train_dataset)

# MobileNetV2 mit vortrainierten Gewichten laden
model_mnv2 = mobilenet_v2(weights=MobileNet_V2_Weights.DEFAULT)
model_mnv2.classifier[1] = nn.Linear(model_mnv2.classifier[1].in_features, len(train_dataset.classes))
model_mnv2 = model_mnv2.to(device)

criterion_mnv2 = nn.CrossEntropyLoss(weight=class_weights)
optimizer_mnv2 = torch.optim.Adam(model_mnv2.parameters(), lr=lr)

print("MobileNetV2 bereit.")
```

```python
best_val_loss_mnv2 = float("inf")
best_epoch_mnv2 = 0
counter_mnv2 = 0
best_path_mnv2 = output_dir / "best_model_mobilenetv2_background.pth"

start_time = perf_counter()

for epoch in range(num_epochs):
    train_loss, train_acc = run_epoch(train_loader, model_mnv2, criterion_mnv2, optimizer_mnv2, train=True)
    val_loss, val_acc = run_epoch(val_loader, model_mnv2, criterion_mnv2, train=False)

    if val_loss < best_val_loss_mnv2 - min_delta:
        best_val_loss_mnv2 = val_loss
        best_epoch_mnv2 = epoch + 1
        counter_mnv2 = 0
        torch.save(model_mnv2.state_dict(), best_path_mnv2)
    else:
        counter_mnv2 += 1

    print(
        f"Epoch {epoch+1}/{num_epochs} | "
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )

    if counter_mnv2 >= patience:
        print(f"Early stopping at epoch {epoch+1}")
        break

elapsed_time_mnv2 = perf_counter() - start_time
print(f"\nTrainingszeit MobileNetV2: {elapsed_time_mnv2:.1f}s")

model_mnv2.load_state_dict(torch.load(best_path_mnv2, map_location=device))
model_mnv2 = model_mnv2.to(device)
```

```python
loss_p_mnv2, acc_p_mnv2, f1_p_mnv2, f1_per_class_p_mnv2, cm_p_mnv2, report_p_mnv2 = evaluate_test(
    test_phacelia_loader, test_phacelia_dataset, model_mnv2, criterion_mnv2, class_names, "Phacelia (MobileNetV2)")

loss_o_mnv2, acc_o_mnv2, f1_o_mnv2, f1_per_class_o_mnv2, cm_o_mnv2, report_o_mnv2 = evaluate_test(
    test_others_loader, test_others_dataset, model_mnv2, criterion_mnv2, class_names, "Others (MobileNetV2)")

print("\n" + "="*50)
print("VERGLEICH MobileNetV2")
print("="*50)
print(f"Phacelia → Accuracy: {acc_p_mnv2:.2f}% | Macro F1: {f1_p_mnv2:.4f}")
print(f"Others   → Accuracy: {acc_o_mnv2:.2f}% | Macro F1: {f1_o_mnv2:.4f}")
print(f"Differenz: {acc_p_mnv2 - acc_o_mnv2:.2f} Prozentpunkte")
```
